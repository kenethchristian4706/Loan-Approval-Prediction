```python
from sklearn.preprocessing import StandardScaler
```

```python
cols = ['no_of_dependents' , 'income_annum' , 'loan_amount'  , 'loan_term' , 'cibil_score' , 'residential_assets_value' , 'commercial_assets_value' , 'luxury_assets_value' , 'bank_asset_value'  ]
```

```python
scaler = StandardScaler()
```

```python
df[cols] = scaler.fit_transform(df[cols])
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.corr()
```

```python
x = df.drop(['loan_id' , 'loan_status' , 'no_of_dependents' ,'education' , 'self_employed' ] , axis=1)
y=df['loan_status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score ,classification_report  , confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
models = {
    "Logistic Reg":LogisticRegression(),
    "KNN":KNeighborsClassifier(n_neighbors=5),
    "Naive_Bayes" : GaussianNB(),
    "Dtree":DecisionTreeClassifier(), 
    "SVM": SVC(kernel='rbf')
}
```

```python
result = []
```

```python
for model_name , model in models.items():
    model.fit(x_train , y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test , y_pred)
    report = classification_report(y_test , y_pred)
    result.append({model_name : [accuracy]})
    print(model_name , report)
    
```
